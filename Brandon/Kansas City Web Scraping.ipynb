```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#Go to KC url
url = "http://www.city-data.com/city/Kansas-City-Missouri.html"
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
city = soup.find('div', class_="container-fluid").find('h1', class_="city").text
localTime = soup.find('div', class_="container-fluid").find('span', id="ourtime").text
population = soup.find('div', class_="container-fluid").find('section', class_="city-population").text
costOfLiving = soup.find('div', class_="container-fluid").find('section', class_="cost-of-living-index").text
nickNames = soup.find('div', class_="container-fluid").find('section', class_="city-nicknames").text

#cityGuides = soup.find('div', class_="list-group row").text

print(f'{city}')
print(f'Current Local Time: {localTime}')
print(f'{population}')
print(f'{costOfLiving}')
print(f'{nickNames}')

#print(f'{cityGuides}')
```
